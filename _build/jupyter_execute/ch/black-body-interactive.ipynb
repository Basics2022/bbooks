```python
import sys
sys.path.append('../')

import numpy as np
import matplotlib.pyplot as plt

import utils.black_body as bb

```

```python
# Wavelength range from 100 nm to 3000 nm
wavelengths = np.linspace(100e-9, 3000e-9, 1000)  # Convert nm to meters

temp = 5000

# Initialize the figure and axes for 3 subplots
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Planck's Law spectrum
intensity = bb.planck(wavelengths, temp)
ax[0].plot(wavelengths * 1e9, intensity, label=f'Temperature: {temp} K')


```
